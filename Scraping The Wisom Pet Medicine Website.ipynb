```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://www.wisdompetmed.com/"
```

```python
res = requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text)
print(soup.prettify())
```

```python
# web_spans = soup.find_all("span")
# for web_span in web_spans:
#     print(web_span.text)
```

```python
phones = soup.find_all('span', class_ ='phone')
for phone in phones:
    print(phone.text)
```

```python
testimonials = soup.find_all('div', class_='quote')
for testimonial in testimonials:
    print(testimonial.text)
```

```python
#find all staff members
staff_members = soup.find_all("div", attrs={'class':'info col-xs-8 col-xs-offset-2 col-sm-7 col-sm-offset-0 col-md-6 col-lg-8'})

doctor_names = []
for staff in staff_members:
    h3_element = staff.find("h3")
    if h3_element:
        doctor_names.append(h3_element.text)
        
for name in doctor_names:
    print(name)
```

```python
#find all links
links = soup.find_all("a")
for link in links:
    print(link.text, link.get('href'))
```

```python
#write code into file
with open('wisdom_vet.txt', 'w') as f:
    f.write(soup.prettify())
```
